```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv("./Data/csv_8.csv")
```

```python
d = d.replace(' - ','0')
d
```

```python
d["합계"] = d.합계.str.replace(',', '').astype('int64')
```

```python
d["수력"] = d.수력.str.replace(',', '').astype('int64')
```

```python
d["기력"] = d.기력.str.replace(',', '').astype('int64')
```

```python
d["복합화력"] = d.복합화력.str.replace(',', '').astype('int64')
```

```python
d["원자력"] = d.원자력.str.replace(',', '').astype('int64')
```

```python
d["신재생"] = d.신재생.str.replace(',', '').astype('int64')
```

```python
d["집단"] = d.집단.str.replace(',', '').astype('int64')
```

```python
d["내연력"] = d.내연력.str.replace(',', '').astype('int64')
```

```python
d["기타"] = d.기타.str.replace(',', '').astype('int64')
```

```python
d["Unnamed: 0"] = d["Unnamed: 0"].astype('string')
```

```python
d
```

```python
x = d['Unnamed: 0'][1:]
a = d['합계'][1:]
b = d["기력"][1:]
c = d["복합화력"][1:]
dd = d["원자력"][1:]
e = d["신재생"][1:]
f = d["집단"][1:]
g = d["내연력"][1:]
h = d["기타"][1:]


plt.figure(figsize=(10,5))
a = [a,b,c,dd,e,f,g,h]
for i in a:
    plt.plot(x, i)
plt.show()
```
